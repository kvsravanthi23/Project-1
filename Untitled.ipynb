```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Now loading the Zomato data file which is in csv format.

```python

dataframe = pd.read_csv("Zomato data .csv")
print(dataframe.head())
```

```python
dataframe.rename(columns={'name':'Restaurant_Name','rate':'rating','listed_in(type)':'Type_of_Restaurant'},inplace=True)
dataframe.info()
```

Before proceeding, let’s convert the data type of the “rate” column to float and remove the denominator.

```python
def handleRate(value):
	value=str(value).split('/')
	value=value[0];
	return float(value)

dataframe['rating']=dataframe['rating'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

lets explore each and every column and get the conslusions

1.exploring types of restuarant column.

```python
sns.countplot(x=dataframe['Type_of_Restaurant'])
plt.xlabel("Type of restaurant")
```

**Conclusion**:The majority of the restaurants fall into the dining category.



Now let us plot a graph between votes and types of resturants using pyplot in matplot library


```python
grouped_data = dataframe.groupby('Type_of_Restaurant')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker="o")
plt.xlabel("Type of restaurant", c="red", size=20)
plt.ylabel("Votes", c="red", size=20)
```

**Conclusion**: Dining restaurants are preferred by a larger number of individuals.

Now we will determine the restaurant’s name that received the maximum votes based on a given dataframe.


```python
max_votes = dataframe['votes'].max()
restaurant_with_max_votes = dataframe.loc[dataframe['votes'] == max_votes, 'Restaurant_Name']

print("Restaurant(s) with the maximum votes:")
print(restaurant_with_max_votes)
```

**Conclusion**:Empire Resturant has received maximum votes .

2.Let’s explore the online_order column.
We are using countplot to  represent the occurrence(counts) of the observation present in the categorical variable.

```python

sns.countplot(x=dataframe['online_order'])
```

**Conclusion**: This suggests that a majority of the restaurants do not accept online orders.



3.Let’s explore the rate column.
We are using histogram because we have continous values ranging from 2 to 4.5 that is rating is done in interval (1-5.0)

Histogram:To summarize discrete or continuous data that are measured on an interval scale

```python
plt.hist(dataframe['rating'],bins=5)
plt.title("Ratings Distribution")
plt.show()
```

**Conclusion**: The majority of restaurants received ratings ranging from 3.5 to 4.

4.Let’s explore the approx_cost(for two people) column.


```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data)
```

**Conclusion**: The majority of couples prefer restaurants with an approximate cost of 300 rupees.



5.Exploring online orders colums with rating column.

Now we will examine whether online orders receive higher ratings than offline orders.

```python
plt.figure(figsize = (6,6))
sns.boxplot(x = 'online_order', y = 'rating', data = dataframe)
```

**Conclusion**: Offline orders received lower ratings in comparison to online orders, which obtained excellent ratings.

Now we draw the heatMap by summarizing all the above conclusions  and get the final output,

```python
pivot_table = dataframe.pivot_table(index='Type_of_Restaurant', columns='online_order', aggfunc='size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Type_of_Restaurant")
plt.show()
```

In above HeatMap the Dining retaurants with no online orders has more dark color(77) which indicates customers are trying to avoid online orders in dining resturants and cafes which has the online ordering has relatively less dark color(15) than comapred to dining with online ordering.

**Conclusion**: Dining restaurants primarily accept offline orders, whereas cafes primarily receive online orders. This suggests that clients prefer to place orders in person at restaurants, but prefer online ordering at cafes.
